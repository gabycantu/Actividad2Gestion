```python
dfch["name"]= dfch["name"].fillna(method="bfill") 
dfch
```

```python
dfch["neighbourhood_group"]= dfch["neighbourhood_group"].fillna(method="bfill") 
dfch
```

```python
dfch["host_name"]= dfch["host_name"].fillna(method="bfill") 
dfch
```

```python
dfch["last_review"]= dfch["last_review"].fillna("31/12/2022") 
dfch
```

```python
dfch["reviews_per_month"]=dfch["reviews_per_month"].fillna(round(dfch["reviews_per_month"].mean(),2))
dfch
```
